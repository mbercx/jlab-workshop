# Octave Demo

Although using MATLAB's (proprietary) software in a Jupyter notebook isn't possible, [Octave](https://www.gnu.org/software/octave/) is a perfectly viable alternative that is distributed freely under the [GNU General Public Licence](https://www.gnu.org/copyleft/gpl.html) (GPL). Below are some typical MATLAB commands (taken from the MATLAB website) that demonstrate the functionality of the notebook:

````octave
x = -2:0.25:2;
y = x;
[X,Y] = meshgrid(x);
````

````octave
F = X.*exp(-X.^2-Y.^2);
surf(X,Y,F)
````

````octave
x = linspace(0,3*pi,200);
y = cos(x) + rand(1,200);  
scatter(x,y)
````

# Installing Octave for JupyterLab

In order to use Octave in JupyterLab, you first need to install it from the [website](https://www.gnu.org/software/octave/download.html). Next, you need to install the octave_kernel package, either using the Anaconda GUI or using the command line:

```
conda install -c conda-forge octave_kernel 
```

You will have to reload the JupyterLab page to be able to start an Octave notebook.
